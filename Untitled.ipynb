```python
# NBA Player Statistics Machine Learning Analysis
# Comprehensive guide to ML models for NBA dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
url = "https://raw.githubusercontent.com/xtrship/CSV-READ/refs/heads/main/nba_stats_clean_final.csv"
df = pd.read_csv('nba_stats_clean_final.csv')

print("Dataset Shape:", df.shape)
print("\nFirst few rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nMissing values:")
print(df.isnull().sum())

# Basic statistics
print("\nBasic Statistics:")
print(df.describe())

# =============================================================================
# DATA PREPROCESSING
# =============================================================================

# Clean the data
print("\n" + "="*50)
print("DATA PREPROCESSING")
print("="*50)

# Handle missing salary values (common in sports datasets)
df['salary'].fillna(0, inplace=True)

# Remove rows with too many missing values
df = df.dropna(subset=['pts_per_g', 'mp_per_g', 'age'])

# Create categorical variables
df['position_category'] = df['pos'].map({'PG': 'Guard', 'SG': 'Guard', 'SF': 'Forward', 
                                        'PF': 'Forward', 'C': 'Center'})

# Create performance tiers based on points per game
df['performance_tier'] = pd.cut(df['pts_per_g'], 
                               bins=[0, 5, 10, 15, 20, float('inf')], 
                               labels=['Bench', 'Role Player', 'Solid Starter', 'Star', 'Superstar'])

# Feature engineering
df['total_rebounds'] = df['orb_per_g'] + df['drb_per_g']
df['efficiency'] = df['pts_per_g'] / (df['fga_per_g'] + 0.1)  # Avoid division by zero
df['usage_rate'] = (df['fga_per_g'] + 0.44 * df['fta_per_g'] + df['tov_per_g']) / df['mp_per_g']

print("Data preprocessing completed!")
print("Final dataset shape:", df.shape)

# =============================================================================
# EXPLORATORY DATA ANALYSIS
# =============================================================================

print("\n" + "="*50)
print("EXPLORATORY DATA ANALYSIS")
print("="*50)

# Correlation matrix for numerical features
numeric_cols = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of NBA Statistics')
plt.tight_layout()
plt.show()

# Distribution of key statistics
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
key_stats = ['pts_per_g', 'ast_per_g', 'trb_per_g', 'fg_pct', 'age', 'mp_per_g']

for i, stat in enumerate(key_stats):
    row, col = i // 3, i % 3
    axes[row, col].hist(df[stat].dropna(), bins=30, alpha=0.7)
    axes[row, col].set_title(f'Distribution of {stat}')
    axes[row, col].set_xlabel(stat)
    axes[row, col].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# =============================================================================
# MACHINE LEARNING MODELS
# =============================================================================

print("\n" + "="*50)
print("MACHINE LEARNING MODELS")
print("="*50)

# Prepare features for modeling
feature_columns = ['age', 'games', 'mp_per_g', 'fg_pct', 'fg3_pct', 'ft_pct', 
                  'orb_per_g', 'drb_per_g', 'ast_per_g', 'stl_per_g', 
                  'blk_per_g', 'tov_per_g', 'pf_per_g']

# Remove rows with missing values in key features
df_clean = df.dropna(subset=feature_columns + ['pts_per_g'])

X = df_clean[feature_columns]
y_regression = df_clean['pts_per_g']  # Target for regression
y_classification = df_clean['performance_tier']  # Target for classification

print(f"Clean dataset shape: {df_clean.shape}")
print(f"Features shape: {X.shape}")

# =============================================================================
# MODEL 1: LINEAR REGRESSION - PREDICTING POINTS PER GAME
# =============================================================================

print("\n" + "-"*30)
print("MODEL 1: LINEAR REGRESSION")
print("-"*30)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_regression, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)

# Predictions
y_pred_lr = lr_model.predict(X_test_scaled)

# Evaluate
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print(f"Linear Regression Results:")
print(f"Mean Squared Error: {mse_lr:.3f}")
print(f"R² Score: {r2_lr:.3f}")

# Feature importance
feature_importance_lr = pd.DataFrame({
    'feature': feature_columns,
    'coefficient': lr_model.coef_
}).sort_values('coefficient', key=abs, ascending=False)

print("\nTop 5 Most Important Features:")
print(feature_importance_lr.head())

# =============================================================================
# MODEL 2: RANDOM FOREST REGRESSION - PREDICTING POINTS PER GAME
# =============================================================================

print("\n" + "-"*30)
print("MODEL 2: RANDOM FOREST REGRESSION")
print("-"*30)

# Train Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Random Forest Results:")
print(f"Mean Squared Error: {mse_rf:.3f}")
print(f"R² Score: {r2_rf:.3f}")

# Feature importance
feature_importance_rf = pd.DataFrame({
    'feature': feature_columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nTop 5 Most Important Features:")
print(feature_importance_rf.head())

# =============================================================================
# MODEL 3: GRADIENT BOOSTING - PREDICTING POINTS PER GAME
# =============================================================================

print("\n" + "-"*30)
print("MODEL 3: GRADIENT BOOSTING REGRESSION")
print("-"*30)

# Train Gradient Boosting
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)

# Predictions
y_pred_gb = gb_model.predict(X_test)

# Evaluate
mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

print(f"Gradient Boosting Results:")
print(f"Mean Squared Error: {mse_gb:.3f}")
print(f"R² Score: {r2_gb:.3f}")

# =============================================================================
# MODEL 4: CLASSIFICATION - PREDICTING PERFORMANCE TIER
# =============================================================================

print("\n" + "-"*30)
print("MODEL 4: RANDOM FOREST CLASSIFICATION")
print("-"*30)

# Prepare classification data
df_classification = df_clean.dropna(subset=['performance_tier'])
X_class = df_classification[feature_columns]
y_class = df_classification['performance_tier']

# Split data
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_class, y_class, test_size=0.2, random_state=42, stratify=y_class)

# Train Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_class, y_train_class)

# Predictions
y_pred_class = rf_classifier.predict(X_test_class)

# Evaluate
accuracy = rf_classifier.score(X_test_class, y_test_class)
print(f"Classification Accuracy: {accuracy:.3f}")

print("\nClassification Report:")
print(classification_report(y_test_class, y_pred_class))

# =============================================================================
# MODEL 5: CLUSTERING - PLAYER ARCHETYPES
# =============================================================================

print("\n" + "-"*30)
print("MODEL 5: K-MEANS CLUSTERING")
print("-"*30)

# Select features for clustering (per-game stats)
cluster_features = ['pts_per_g', 'ast_per_g', 'trb_per_g', 'stl_per_g', 'blk_per_g', 'fg_pct']
X_cluster = df[cluster_features].dropna()

# Scale features
scaler_cluster = StandardScaler()
X_cluster_scaled = scaler_cluster.fit_transform(X_cluster)

# Find optimal number of clusters
inertias = []
K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_cluster_scaled)
    inertias.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Apply K-means with optimal clusters (let's use 5 for player archetypes)
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_cluster_scaled)

# Analyze clusters
cluster_df = X_cluster.copy()
cluster_df['cluster'] = cluster_labels

print("Cluster Analysis:")
for i in range(5):
    cluster_data = cluster_df[cluster_df['cluster'] == i]
    print(f"\nCluster {i} ({len(cluster_data)} players):")
    print(cluster_data[cluster_features].mean().round(2))

# =============================================================================
# MODEL 6: PRINCIPAL COMPONENT ANALYSIS
# =============================================================================

print("\n" + "-"*30)
print("MODEL 6: PRINCIPAL COMPONENT ANALYSIS")
print("-"*30)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_cluster_scaled)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 'bo-')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('PCA - Cumulative Explained Variance')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Variance')
plt.legend()
plt.show()

print(f"Components needed for 95% variance: {np.argmax(cumulative_variance >= 0.95) + 1}")

# =============================================================================
# MODEL COMPARISON AND RECOMMENDATIONS
# =============================================================================

print("\n" + "="*50)
print("MODEL COMPARISON AND RECOMMENDATIONS")
print("="*50)

# Compare regression models
results_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest', 'Gradient Boosting'],
    'MSE': [mse_lr, mse_rf, mse_gb],
    'R² Score': [r2_lr, r2_rf, r2_gb]
})

print("Regression Model Comparison:")
print(results_df)

print(f"\nBest performing model: {results_df.loc[results_df['R² Score'].idxmax(), 'Model']}")

# =============================================================================
# ADVANCED MODEL EXAMPLES
# =============================================================================

print("\n" + "-"*30)
print("ADVANCED MODELS")
print("-"*30)

# Regularized regression (Ridge)
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled, y_train)
y_pred_ridge = ridge_model.predict(X_test_scaled)
r2_ridge = r2_score(y_test, y_pred_ridge)

print(f"Ridge Regression R²: {r2_ridge:.3f}")

# Lasso regression
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train_scaled, y_train)
y_pred_lasso = lasso_model.predict(X_test_scaled)
r2_lasso = r2_score(y_test, y_pred_lasso)

print(f"Lasso Regression R²: {r2_lasso:.3f}")

# =============================================================================
# SPECIFIC USE CASES AND MODEL RECOMMENDATIONS
# =============================================================================

print("\n" + "="*50)
print("MODEL RECOMMENDATIONS BY USE CASE")
print("="*50)

print("""
1. PREDICTING PLAYER PERFORMANCE (Points, Assists, Rebounds):
   - Best Models: Random Forest, Gradient Boosting, XGBoost
   - Why: Handle non-linear relationships, feature interactions
   - Target Variables: pts_per_g, ast_per_g, trb_per_g

2. PLAYER CLASSIFICATION (Position, Performance Tier):
   - Best Models: Random Forest Classifier, SVM, Logistic Regression
   - Why: Good for categorical predictions
   - Target Variables: pos, performance_tier

3. SALARY PREDICTION:
   - Best Models: Random Forest, Gradient Boosting, Neural Networks
   - Why: Complex relationships between performance and salary
   - Target Variable: salary

4. PLAYER SIMILARITY/CLUSTERING:
   - Best Models: K-Means, Hierarchical Clustering
   - Why: Identify player archetypes and similar players
   - Features: All per-game statistics

5. INJURY RISK PREDICTION:
   - Best Models: Logistic Regression, Random Forest
   - Why: Good for binary classification
   - Target: games_played vs expected

6. DRAFT PROSPECT EVALUATION:
   - Best Models: Ensemble methods (Random Forest + Gradient Boosting)
   - Why: Multiple aspects of player evaluation
   - Features: Age-adjusted statistics

7. TRADE VALUE ASSESSMENT:
   - Best Models: Multi-output regression
   - Why: Multiple factors determine trade value
   - Targets: salary, performance metrics
""")

# =============================================================================
# FEATURE IMPORTANCE VISUALIZATION
# =============================================================================

print("\n" + "-"*30)
print("FEATURE IMPORTANCE ANALYSIS")
print("-"*30)

# Plot feature importance from Random Forest
plt.figure(figsize=(12, 8))
feature_importance_rf_sorted = feature_importance_rf.sort_values('importance')
plt.barh(feature_importance_rf_sorted['feature'], feature_importance_rf_sorted['importance'])
plt.xlabel('Feature Importance')
plt.title('Random Forest Feature Importance for Points Prediction')
plt.tight_layout()
plt.show()

# =============================================================================
# MODEL VALIDATION AND HYPERPARAMETER TUNING EXAMPLE
# =============================================================================

print("\n" + "-"*30)
print("HYPERPARAMETER TUNING EXAMPLE")
print("-"*30)

# Grid search for Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(RandomForestRegressor(random_state=42), 
                          param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_:.3f}")

# =============================================================================
# PRACTICAL EXAMPLES WITH YOUR DATA
# =============================================================================

print("\n" + "="*50)
print("PRACTICAL EXAMPLES")
print("="*50)

# Example 1: Find players similar to a specific player
def find_similar_players(player_name, df, top_n=5):
    """Find players most similar to the given player based on stats"""
    player_stats = df[df['player_name'] == player_name][feature_columns].values
    if len(player_stats) == 0:
        return "Player not found"
    
    # Calculate distances
    distances = []
    for idx, row in df.iterrows():
        if row['player_name'] != player_name:
            stats = row[feature_columns].values
            if not np.isnan(stats).any():
                distance = np.sqrt(np.sum((player_stats[0] - stats)**2))
                distances.append((row['player_name'], distance))
    
    # Sort and return top N
    distances.sort(key=lambda x: x[1])
    return distances[:top_n]

# Example usage
sample_player = df['player_name'].iloc[100]  # Get a sample player
similar_players = find_similar_players(sample_player, df_clean)
print(f"\nPlayers similar to {sample_player}:")
for player, distance in similar_players:
    print(f"  {player}: Distance = {distance:.2f}")

# Example 2: Predict a player's points based on other stats
def predict_player_points(age, games, mp_per_g, fg_pct, fg3_pct, ft_pct, 
                         orb_per_g, drb_per_g, ast_per_g, stl_per_g, 
                         blk_per_g, tov_per_g, pf_per_g):
    """Predict points per game for a player with given stats"""
    player_stats = np.array([[age, games, mp_per_g, fg_pct, fg3_pct, ft_pct,
                             orb_per_g, drb_per_g, ast_per_g, stl_per_g,
                             blk_per_g, tov_per_g, pf_per_g]])
    player_stats_scaled = scaler.transform(player_stats)
    predicted_points = rf_model.predict(player_stats)[0]
    return predicted_points

# Example prediction
sample_prediction = predict_player_points(25, 70, 30, 0.45, 0.35, 0.80, 
                                        1.5, 4.0, 5.0, 1.0, 0.5, 2.0, 2.5)
print(f"\nSample prediction: {sample_prediction:.1f} points per game")

print("\n" + "="*50)
print("NEXT STEPS AND MODEL SELECTION GUIDE")
print("="*50)

print("""
CHOOSING THE RIGHT MODEL:

1. **For Point Prediction**: Random Forest or Gradient Boosting
   - Handles non-linear relationships well
   - Good feature importance interpretation

2. **For Player Classification**: Random Forest Classifier
   - Robust to outliers
   - Handles mixed data types well

3. **For Interpretability**: Linear/Logistic Regression
   - Clear coefficient interpretation
   - Fast training and prediction

4. **For Complex Patterns**: Neural Networks (requires more data)
   - Can capture complex non-linear relationships
   - Requires careful tuning

5. **For Player Grouping**: K-Means Clustering
   - Identifies natural player archetypes
   - Useful for scouting and team building

VALIDATION STRATEGIES:
- Use cross-validation for small datasets
- Time-based splits for temporal data
- Stratified sampling for imbalanced classes

FEATURE ENGINEERING IDEAS:
- Advanced metrics (PER, True Shooting %, Usage Rate)
- Rolling averages over multiple games
- Age-adjusted performance metrics
- Team-level statistics
- Positional adjustments
""")
```

```python
# NBA Position Clustering by Assists and Playmaking
# Specialized clustering model to group positions by assist patterns

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
url = "https://raw.githubusercontent.com/xtrship/CSV-READ/refs/heads/main/nba_stats_clean_final.csv"
df = pd.read_csv(url)

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")

# =============================================================================
# DATA PREPARATION FOR ASSIST-BASED CLUSTERING
# =============================================================================

print("\n" + "="*60)
print("DATA PREPARATION FOR ASSIST-BASED CLUSTERING")
print("="*60)

# Select playmaking-related features
playmaking_features = [
    'ast_per_g',      # Primary assists
    'tov_per_g',      # Turnovers (playmaking cost)
    'mp_per_g',       # Minutes (opportunity)
    'fg_pct',         # Shooting efficiency
    'pts_per_g',      # Scoring ability
    'stl_per_g'       # Defensive playmaking
]

# Create assist-to-turnover ratio
df['ast_tov_ratio'] = df['ast_per_g'] / (df['tov_per_g'] + 0.1)  # Avoid division by zero

# Create assist rate (assists per minute)
df['ast_rate'] = df['ast_per_g'] / (df['mp_per_g'] + 0.1)

# Add these new features
extended_features = playmaking_features + ['ast_tov_ratio', 'ast_rate']

# Clean data - remove rows with missing values in key features
df_clean = df.dropna(subset=extended_features + ['pos'])

# Filter for players with meaningful minutes (>10 mpg)
df_clean = df_clean[df_clean['mp_per_g'] >= 10]

print(f"Clean dataset shape: {df_clean.shape}")
print(f"Positions in dataset: {df_clean['pos'].value_counts()}")

# =============================================================================
# EXPLORATORY ANALYSIS: ASSISTS BY POSITION
# =============================================================================

print("\n" + "-"*40)
print("ASSISTS ANALYSIS BY POSITION")
print("-"*40)

# Assists by position analysis
position_assist_stats = df_clean.groupby('pos').agg({
    'ast_per_g': ['mean', 'median', 'std', 'min', 'max'],
    'ast_tov_ratio': ['mean', 'median'],
    'ast_rate': ['mean', 'median']
}).round(2)

print("Assist Statistics by Position:")
print(position_assist_stats)

# Visualization: Assists by Position
plt.figure(figsize=(15, 10))

# Subplot 1: Box plot of assists by position
plt.subplot(2, 2, 1)
sns.boxplot(data=df_clean, x='pos', y='ast_per_g')
plt.title('Assists per Game by Position')
plt.xticks(rotation=45)

# Subplot 2: Assist-to-Turnover ratio by position
plt.subplot(2, 2, 2)
sns.boxplot(data=df_clean, x='pos', y='ast_tov_ratio')
plt.title('Assist-to-Turnover Ratio by Position')
plt.xticks(rotation=45)

# Subplot 3: Assist rate by position
plt.subplot(2, 2, 3)
sns.boxplot(data=df_clean, x='pos', y='ast_rate')
plt.title('Assist Rate (Assists per Minute) by Position')
plt.xticks(rotation=45)

# Subplot 4: Scatter plot of assists vs turnovers
plt.subplot(2, 2, 4)
colors = {'PG': 'red', 'SG': 'blue', 'SF': 'green', 'PF': 'orange', 'C': 'purple'}
for pos in df_clean['pos'].unique():
    pos_data = df_clean[df_clean['pos'] == pos]
    plt.scatter(pos_data['ast_per_g'], pos_data['tov_per_g'], 
                c=colors.get(pos, 'black'), label=pos, alpha=0.6)
plt.xlabel('Assists per Game')
plt.ylabel('Turnovers per Game')
plt.title('Assists vs Turnovers by Position')
plt.legend()

plt.tight_layout()
plt.show()

# =============================================================================
# CLUSTERING MODEL: GROUP POSITIONS BY ASSIST PATTERNS
# =============================================================================

print("\n" + "="*60)
print("CLUSTERING MODEL: POSITION GROUPS BY ASSIST PATTERNS")
print("="*60)

# Prepare clustering features
X_clustering = df_clean[extended_features].copy()

# Handle any remaining missing values
X_clustering = X_clustering.fillna(X_clustering.mean())

print("Clustering features:")
print(X_clustering.describe())

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clustering)

# Find optimal number of clusters using elbow method and silhouette score
print("\nFinding optimal number of clusters...")

inertias = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, cluster_labels))

# Plot elbow curve and silhouette scores
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Elbow plot
ax1.plot(K_range, inertias, 'bo-')
ax1.set_xlabel('Number of Clusters (k)')
ax1.set_ylabel('Inertia')
ax1.set_title('Elbow Method for Optimal k')
ax1.grid(True)

# Silhouette score plot
ax2.plot(K_range, silhouette_scores, 'ro-')
ax2.set_xlabel('Number of Clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score vs Number of Clusters')
ax2.grid(True)

plt.tight_layout()
plt.show()

# Choose optimal number of clusters (highest silhouette score)
optimal_k = K_range[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters: {optimal_k}")
print(f"Best silhouette score: {max(silhouette_scores):.3f}")

# =============================================================================
# APPLY CLUSTERING WITH OPTIMAL K
# =============================================================================

print("\n" + "-"*40)
print("APPLYING CLUSTERING MODEL")
print("-"*40)

# Apply K-means with optimal number of clusters
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_scaled)

# Add cluster labels to dataframe
df_clustered = df_clean.copy()
df_clustered['assist_cluster'] = cluster_labels

print(f"Clustering completed with {optimal_k} clusters")

# =============================================================================
# ANALYZE CLUSTERS
# =============================================================================

print("\n" + "-"*40)
print("CLUSTER ANALYSIS")
print("-"*40)

# Analyze each cluster
cluster_analysis = df_clustered.groupby('assist_cluster').agg({
    'ast_per_g': ['mean', 'median', 'std'],
    'ast_tov_ratio': ['mean', 'median'],
    'ast_rate': ['mean', 'median'],
    'pts_per_g': ['mean', 'median'],
    'mp_per_g': ['mean', 'median'],
    'pos': lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else 'Mixed'
}).round(2)

print("Cluster Analysis Summary:")
print(cluster_analysis)

# Position distribution within each cluster
print("\nPosition Distribution by Cluster:")
position_distribution = pd.crosstab(df_clustered['assist_cluster'], 
                                   df_clustered['pos'], 
                                   normalize='index') * 100
print(position_distribution.round(1))

# =============================================================================
# CLUSTER INTERPRETATION
# =============================================================================

print("\n" + "-"*40)
print("CLUSTER INTERPRETATION")
print("-"*40)

# Define cluster names based on assist patterns
cluster_names = {}
cluster_descriptions = {}

for cluster_id in range(optimal_k):
    cluster_data = df_clustered[df_clustered['assist_cluster'] == cluster_id]
    avg_assists = cluster_data['ast_per_g'].mean()
    avg_ratio = cluster_data['ast_tov_ratio'].mean()
    dominant_position = cluster_data['pos'].mode().iloc[0] if len(cluster_data['pos'].mode()) > 0 else 'Mixed'
    
    # Name clusters based on assist levels
    if avg_assists >= 6:
        cluster_names[cluster_id] = "Primary Playmakers"
        cluster_descriptions[cluster_id] = "High assist players, typically point guards"
    elif avg_assists >= 3:
        cluster_names[cluster_id] = "Secondary Playmakers"
        cluster_descriptions[cluster_id] = "Moderate assists, versatile players"
    elif avg_assists >= 1.5:
        cluster_names[cluster_id] = "Occasional Playmakers"
        cluster_descriptions[cluster_id] = "Limited playmaking, focus on other skills"
    else:
        cluster_names[cluster_id] = "Non-Playmakers"
        cluster_descriptions[cluster_id] = "Minimal assists, specialized roles"

print("Cluster Interpretations:")
for cluster_id in range(optimal_k):
    cluster_data = df_clustered[df_clustered['assist_cluster'] == cluster_id]
    print(f"\nCluster {cluster_id}: {cluster_names[cluster_id]}")
    print(f"  Description: {cluster_descriptions[cluster_id]}")
    print(f"  Size: {len(cluster_data)} players")
    print(f"  Avg Assists: {cluster_data['ast_per_g'].mean():.1f}")
    print(f"  Avg A/TO Ratio: {cluster_data['ast_tov_ratio'].mean():.1f}")
    print(f"  Dominant Position: {cluster_data['pos'].mode().iloc[0] if len(cluster_data['pos'].mode()) > 0 else 'Mixed'}")
    print(f"  Top Players: {', '.join(cluster_data.nlargest(3, 'ast_per_g')['player_name'].tolist())}")

# =============================================================================
# VISUALIZATION: CLUSTER RESULTS
# =============================================================================

print("\n" + "-"*40)
print("CLUSTER VISUALIZATION")
print("-"*40)

# Create comprehensive visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Assists vs Turnovers colored by cluster
ax1 = axes[0, 0]
scatter = ax1.scatter(df_clustered['ast_per_g'], df_clustered['tov_per_g'], 
                     c=df_clustered['assist_cluster'], cmap='viridis', alpha=0.6)
ax1.set_xlabel('Assists per Game')
ax1.set_ylabel('Turnovers per Game')
ax1.set_title('Clusters by Assists vs Turnovers')
plt.colorbar(scatter, ax=ax1)

# Plot 2: Assist rate vs Minutes played
ax2 = axes[0, 1]
scatter2 = ax2.scatter(df_clustered['mp_per_g'], df_clustered['ast_rate'], 
                      c=df_clustered['assist_cluster'], cmap='viridis', alpha=0.6)
ax2.set_xlabel('Minutes per Game')
ax2.set_ylabel('Assist Rate (Assists/Minute)')
ax2.set_title('Clusters by Minutes vs Assist Rate')
plt.colorbar(scatter2, ax=ax2)

# Plot 3: Position distribution by cluster
ax3 = axes[1, 0]
position_cluster_counts = pd.crosstab(df_clustered['pos'], df_clustered['assist_cluster'])
position_cluster_counts.plot(kind='bar', stacked=True, ax=ax3, colormap='viridis')
ax3.set_xlabel('Position')
ax3.set_ylabel('Number of Players')
ax3.set_title('Position Distribution Across Clusters')
ax3.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.setp(ax3.xaxis.get_majorticklabels(), rotation=45)

# Plot 4: Cluster centers (radar chart style)
ax4 = axes[1, 1]
cluster_centers = df_clustered.groupby('assist_cluster')[extended_features].mean()
cluster_centers_scaled = pd.DataFrame(
    scaler.inverse_transform(kmeans_final.cluster_centers_),
    columns=extended_features
)

# Plot parallel coordinates for cluster centers
for i in range(optimal_k):
    ax4.plot(range(len(extended_features)), cluster_centers_scaled.iloc[i], 
             marker='o', label=f'Cluster {i}: {cluster_names[i]}')

ax4.set_xticks(range(len(extended_features)))
ax4.set_xticklabels(extended_features, rotation=45, ha='right')
ax4.set_ylabel('Feature Value')
ax4.set_title('Cluster Centers - Feature Profiles')
ax4.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# DETAILED POSITION-CLUSTER ANALYSIS
# =============================================================================

print("\n" + "="*60)
print("DETAILED POSITION-CLUSTER ANALYSIS")
print("="*60)

# Create a detailed breakdown
for pos in ['PG', 'SG', 'SF', 'PF', 'C']:
    pos_data = df_clustered[df_clustered['pos'] == pos]
    if len(pos_data) > 0:
        print(f"\n{pos} (n={len(pos_data)}):")
        cluster_dist = pos_data['assist_cluster'].value_counts().sort_index()
        
        for cluster_id in cluster_dist.index:
            count = cluster_dist[cluster_id]
            percentage = (count / len(pos_data)) * 100
            cluster_players = pos_data[pos_data['assist_cluster'] == cluster_id]
            avg_assists = cluster_players['ast_per_g'].mean()
            
            print(f"  Cluster {cluster_id} ({cluster_names[cluster_id]}): "
                  f"{count} players ({percentage:.1f}%) - Avg {avg_assists:.1f} assists")
            
            # Show top players in this cluster for this position
            top_players = cluster_players.nlargest(3, 'ast_per_g')['player_name'].tolist()
            print(f"    Top players: {', '.join(top_players)}")

# =============================================================================
# ADVANCED CLUSTERING: POSITION-SPECIFIC ANALYSIS
# =============================================================================

print("\n" + "="*60)
print("POSITION-SPECIFIC CLUSTERING")
print("="*60)

def cluster_position_by_assists(position, n_clusters=3):
    """
    Cluster players within a specific position by their assist patterns
    """
    pos_data = df_clean[df_clean['pos'] == position].copy()
    
    if len(pos_data) < 10:  # Skip if too few players
        return None
    
    # Features for position-specific clustering
    X_pos = pos_data[extended_features]
    X_pos_scaled = StandardScaler().fit_transform(X_pos)
    
    # Apply clustering
    kmeans_pos = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    pos_clusters = kmeans_pos.fit_predict(X_pos_scaled)
    
    pos_data['pos_assist_cluster'] = pos_clusters
    
    return pos_data, kmeans_pos

# Analyze each position separately
position_results = {}

for position in ['PG', 'SG', 'SF', 'PF', 'C']:
    result = cluster_position_by_assists(position, n_clusters=3)
    if result is not None:
        pos_data, model = result
        position_results[position] = pos_data
        
        print(f"\n{position} Position - Assist-based Sub-clusters:")
        
        for cluster_id in range(3):
            cluster_players = pos_data[pos_data['pos_assist_cluster'] == cluster_id]
            if len(cluster_players) > 0:
                avg_ast = cluster_players['ast_per_g'].mean()
                avg_ratio = cluster_players['ast_tov_ratio'].mean()
                
                # Determine cluster type
                if position == 'PG':
                    if avg_ast >= 7:
                        cluster_type = "Elite Facilitators"
                    elif avg_ast >= 4:
                        cluster_type = "Standard Playmakers"
                    else:
                        cluster_type = "Score-First Guards"
                elif position in ['SG', 'SF']:
                    if avg_ast >= 4:
                        cluster_type = "Playmaking Wings"
                    elif avg_ast >= 2:
                        cluster_type = "Versatile Players"
                    else:
                        cluster_type = "Scoring Specialists"
                else:  # PF, C
                    if avg_ast >= 3:
                        cluster_type = "Playmaking Bigs"
                    elif avg_ast >= 1.5:
                        cluster_type = "Well-rounded Bigs"
                    else:
                        cluster_type = "Traditional Bigs"
                
                print(f"  Cluster {cluster_id} - {cluster_type}:")
                print(f"    Players: {len(cluster_players)}")
                print(f"    Avg Assists: {avg_ast:.1f}")
                print(f"    Avg A/TO Ratio: {avg_ratio:.1f}")
                
                # Top 3 players in this cluster
                top_players = cluster_players.nlargest(3, 'ast_per_g')
                print(f"    Top Players: {', '.join(top_players['player_name'].tolist())}")

# =============================================================================
# PCA VISUALIZATION OF ASSIST CLUSTERS
# =============================================================================

print("\n" + "-"*40)
print("PCA VISUALIZATION")
print("-"*40)

# Apply PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot PCA results
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='viridis', alpha=0.6)
plt.colorbar(scatter)
plt.xlabel(f'First Principal Component ({pca.explained_variance_ratio_[0]:.1%} variance)')
plt.ylabel(f'Second Principal Component ({pca.explained_variance_ratio_[1]:.1%} variance)')
plt.title('PCA Visualization of Assist-based Clusters')

# Add cluster centers
centers_pca = pca.transform(kmeans_final.cluster_centers_)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', marker='x', s=200, linewidths=3)

plt.show()

print(f"PCA explains {pca.explained_variance_ratio_.sum():.1%} of the variance with 2 components")

# =============================================================================
# PRACTICAL APPLICATIONS
# =============================================================================

print("\n" + "="*60)
print("PRACTICAL APPLICATIONS")
print("="*60)

def find_playmaking_outliers(df_clustered):
    """Find players who are playmaking outliers for their position"""
    outliers = []
    
    for pos in ['SG', 'SF', 'PF', 'C']:  # Non-PG positions
        pos_data = df_clustered[df_clustered['pos'] == pos]
        if len(pos_data) > 0:
            # Find players with unusually high assists for their position
            pos_median_assists = pos_data['ast_per_g'].median()
            high_assist_threshold = pos_data['ast_per_g'].quantile(0.9)
            
            high_assist_players = pos_data[pos_data['ast_per_g'] >= high_assist_threshold]
            
            for _, player in high_assist_players.iterrows():
                outliers.append({
                    'player': player['player_name'],
                    'position': player['pos'],
                    'assists': player['ast_per_g'],
                    'cluster': player['assist_cluster'],
                    'cluster_name': cluster_names[player['assist_cluster']]
                })
    
    return pd.DataFrame(outliers)

outliers_df = find_playmaking_outliers(df_clustered)
print("Playmaking Outliers (Non-guards with high assists):")
print(outliers_df.head(10))

# =============================================================================
# PLAYER LOOKUP FUNCTION
# =============================================================================

def analyze_player_playmaking(player_name, df_clustered):
    """Analyze a specific player's playmaking profile"""
    player_data = df_clustered[df_clustered['player_name'] == player_name]
    
    if len(player_data) == 0:
        return f"Player '{player_name}' not found in dataset"
    
    player = player_data.iloc[0]
    cluster_id = player['assist_cluster']
    
    result = f"""
Player: {player['player_name']}
Position: {player['pos']}
Season: {player['season']}

PLAYMAKING PROFILE:
- Assists per Game: {player['ast_per_g']:.1f}
- Assist-to-Turnover Ratio: {player['ast_tov_ratio']:.1f}
- Assist Rate: {player['ast_rate']:.3f}
- Cluster: {cluster_id} ({cluster_names[cluster_id]})

COMPARISON TO POSITION:
"""
    
    # Compare to position averages
    pos_avg = df_clustered[df_clustered['pos'] == player['pos']]['ast_per_g'].mean()
    pos_comparison = "above" if player['ast_per_g'] > pos_avg else "below"
    
    result += f"- {player['ast_per_g']:.1f} assists vs {pos_avg:.1f} position average ({pos_comparison})\n"
    
    # Find similar players in the same cluster
    cluster_players = df_clustered[df_clustered['assist_cluster'] == cluster_id]
    similar_players = cluster_players[cluster_players['player_name'] != player_name].nsmallest(3, 
        abs(cluster_players['ast_per_g'] - player['ast_per_g']))
    
    result += f"\nSimilar Playmaking Style:\n"
    for _, similar in similar_players.iterrows():
        result += f"- {similar['player_name']} ({similar['pos']}, {similar['ast_per_g']:.1f} ast)\n"
    
    return result

# Example player analysis
sample_players = ['LeBron James', 'Chris Paul', 'Nikola Jokic', 'Giannis Antetokounmpo']
available_players = df_clustered['player_name'].unique()

print("\nSample Player Analysis:")
for player in sample_players:
    if player in available_players:
        print(analyze_player_playmaking(player, df_clustered))
        break

# =============================================================================
# EXPORT RESULTS AND RECOMMENDATIONS
# =============================================================================

print("\n" + "="*60)
print("EXPORT RESULTS AND USAGE RECOMMENDATIONS")
print("="*60)

# Create summary dataframe
summary_df = df_clustered[['player_name', 'pos', 'season', 'ast_per_g', 
                          'ast_tov_ratio', 'ast_rate', 'assist_cluster']].copy()
summary_df['cluster_name'] = summary_df['assist_cluster'].map(cluster_names)

print("Summary DataFrame created with cluster assignments")
print(f"Shape: {summary_df.shape}")
print("\nSample of results:")
print(summary_df.head(10))

print("\n" + "="*60)
print("USAGE RECOMMENDATIONS")
print("="*60)

print("""
HOW TO USE THIS CLUSTERING MODEL:

1. **Team Building**: 
   - Identify which clusters your team lacks
   - Find undervalued players in specific clusters
   
2. **Player Development**:
   - Help players transition between clusters
   - Set assist targets based on cluster profiles
   
3. **Scouting**:
   - Compare prospects to established cluster patterns
   - Identify position-switching candidates
   
4. **Trade Analysis**:
   - Find players with similar playmaking profiles
   - Assess fit within team's existing clusters

5. **Tactical Analysis**:
   - Understand team's playmaking distribution
   - Plan rotations based on cluster balance

NEXT STEPS:
1. Run this analysis on current season data
2. Track how players move between clusters over time
3. Correlate clusters with team success metrics
4. Add salary data to find value in each cluster

CUSTOMIZATION OPTIONS:
- Adjust clustering features based on your specific needs
- Change number of clusters for more/less granular groupings
- Apply time-based clustering to see evolution
- Include defensive stats for more complete player profiles
""")

# Function to save results
def save_clustering_results(df_clustered, filename='nba_assist_clusters.csv'):
    """Save clustering results to CSV"""
    export_df = df_clustered[['player_name', 'current_team', 'pos', 'season', 
                             'ast_per_g', 'ast_tov_ratio', 'ast_rate', 
                             'assist_cluster']].copy()
    export_df['cluster_name'] = export_df['assist_cluster'].map(cluster_names)
    export_df.to_csv(filename, index=False)
    print(f"Results saved to {filename}")

# Uncomment to save results
# save_clustering_results(df_clustered)

print("\nClustering analysis complete!")
```
